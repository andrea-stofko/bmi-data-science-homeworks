### Task 5: Simple  Linear Regression 
Use the `ols` function from the [statsmodels](http://www.statsmodels.org/stable/index.html) package to regress the Sold price on some of the other variables. Your model should be of the form:
$$
\text{Sold Price} = \beta_0 + \beta_1 x, 
$$
where $x$ is one of the other variables. 

You'll find that the best predictor of sold price is the listing price. Report the R-squared value for this model (`Sold_Price ~ Listing_Price`) and give an interpretation for its meaning. Also give an interpretation of $\beta_1$ for this model. Make a scatterplot of listing price vs. sold price and overlay the prediction coming from your regression model. 

****Create predictive model for Sold_Price based on List_Price****

````python
import statsmodels.formula.api as sm 

#use ols function to look at predictors -- try a couple others first
#price_pred = sm.ols(formula = "Sold_Price ~ Acres", data = house_data6).fit()
#price_pred = sm.ols(formula = "Sold_Price ~ Total_Bedrooms", data = house_data6).fit()

#Look at Listing_Price as predictor
price_pred = sm.ols(formula = "Sold_Price ~ Listing_Price", data = house_data6).fit()
price_pred.summary()
````

**Your Interpretation:** The R-Squared value for the Sold_Price ~ Listing_Price predictor is 0.944, demonstrating that Listing_Price is a strong indicator of the Sold_Price. B1, or the predicted slope, is 0.92 for this model, meaning that we would expect the Sold Price to be 92% of the Listing Price for a given home listing.

****Create a scatterplot for the sold price versus listed price****

````python
#create a scatterplot of listing price versus sold price
plt.scatter(house_data6['Listing_Price'], house_data6['Sold_Price'], color = 'cornflowerblue')

#create line for predicted values
house_par = price_pred.params #extract intercept and slope
x = house_data6["Listing_Price"] #extract Listed_Price values
y = house_par["Intercept"] + (house_par["Listing_Price"] * x) #predicted sold price is function of listed price
plt.plot(x,y, label = "Predicted Selling Price") #plot and label regression line

#add specifications for plot
plt.title("Listing Price versus Sold Price")
plt.ticklabel_format(style="plain") #get rid of scientific notation for prices
plt.ylabel("Sold Price")
plt.xlabel("Listing Price")
plt.legend()
plt.show()
````
